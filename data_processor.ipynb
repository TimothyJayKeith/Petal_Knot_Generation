```python
# Import modules
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import pickle
import ast
import ipynb.fs.full.petal_assistant as pa
```

# Create a dataset with only the permutations and signa

```python
petal_knot_table = pd.read_csv("finished_petalKnotTable_e2.csv")
petal_knot_table
```

```python
Signatures = petal_knot_table["Signature"]
Permutations = petal_knot_table["permutation"]
Signatures
```

```python
keys = [i for i in range(len(Signatures)) if Signatures[i] in range(-5, 5) and Permutations[i][0] == "["]
```

```python
train_data, test_data = train_test_split(petal_knot_table)
```

```python
train_signatures = train_data["Signature"]
test_signatures = test_data["Signature"]
train_permutations = train_data["permutation"]
test_permutations = test_data["permutation"]
```

```python
train_labels = np.array([train_signatures[i] for i in train_signatures.keys() if i in keys])
```

```python
test_labels = np.array([test_signatures[i] for i in test_signatures.keys() if i in keys])
```

```python
train_features = pa.process_permutation_dataset([train_permutations[i] for i in train_permutations.keys() if i in keys])
```

```python
test_features = pa.process_permutation_dataset([test_permutations[i] for i in test_permutations.keys() if i in keys])
```

```python
small_signature_dataset = (train_features, test_features, train_labels, test_labels)
```

```python
with open("small_signature_dataset.pkl", "wb") as f:
    pickle.dump(small_signature_dataset, f)
```
